```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
THC = yf.download("THC", start="2024-01-29")
```

```python
thc = THC[['Close']]
thc
```

```python
thc["EMA20"]=thc["Close"]["THC"].ewm(span=20,min_periods=20).mean()
thc["EMA50"]=thc["Close"]["THC"].ewm(span=50,min_periods=50).mean()
```

```python
thc.plot(figsize=(12,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
