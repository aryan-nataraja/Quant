```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
THC = yf.download("THC", start="2024-01-29")
```

```python
thc = THC[['Close']]
thc
```

```python
thc.plot(figsize=(12,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
thc["SMA50"]=thc["Close"]["THC"].rolling(window=50,min_periods=50).mean()
```

```python
thc.head(52)
```

```python
thc.plot(figsize=(12,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
thc["SMA200"]=thc["Close"]["THC"].rolling(window=200,min_periods=200).mean()
```

```python
thc.plot(figsize=(12,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
