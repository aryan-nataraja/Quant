```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = ["LLY", "NVDA", "THC", "NVO", "MSFT", "AAPL", "JNJ"]
```

```python
stocks = yf.download(ticker, start = "2015-01-01", end="2025-01-01")
```

```python
stocks
```

```python
close=stocks.loc[:,"Close"].copy()
close
```

```python
close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

# Normalise the close values so stocks are comparable

```python
normclose=close.div(close.iloc[0]).mul(100)
normclose
```

```python
normclose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe().T
```

```python
summary=ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=15)

plt.xlabel("Annual risk(std)", fontsize=15)
plt.ylabel("Annual return", fontsize=15)
plt.title("Risk/return", fontsize=25)
```

# Conclusion

Based on the analysis done above, we can see that NVDA while having high risk has very high return. LLY, MSFT and AAPL have a good balance between risk and reward. 

A basic portfoio of NVDA, AAPL, MSFT and LLY while tech heavy is balanced with some conservative and aggressive stocks. Its diversification is also evident given the not so high correlation scores between the 4 stocks.
